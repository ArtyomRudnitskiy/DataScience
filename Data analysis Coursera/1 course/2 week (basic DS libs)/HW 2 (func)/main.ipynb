# Строим реальную функцию

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy

x = np.linspace(1, 15, 100)
f = np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)

plt.plot(x, f, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

# Приближаем полиномом первой степени (решение СЛАУ)

```python
# Aw=b
# А должно быть известно -> это степени иксов
# w нам неизвестно -> это коэффициенты многочлена w0 и w1
# b - значения функции в точка, т.е. надо посчитать

# пробуем с первой степенью
X = np.array([1, 15])

A = np.c_[np.ones(2), X]
b = np.sin(X/5) * np.exp(X/10) + 5*np.exp(-X/2)

w = np.linalg.solve(A, b)

# y = k*x+b т.е. line1 = w1*x + w0
line1 = w[1]*x + w[0]

plt.plot(x, f, 'r', label='real')
plt.plot(x, line1, 'g', label='approximation')
plt.legend(prop={'size':10})
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

# Приближаем полиномом второй степени

```python
X = np.array([1, 8, 15])

A = np.c_[np.ones(3), X, X**2]
b = np.sin(X/5) * np.exp(X/10) + 5*np.exp(-X/2)

w = np.linalg.solve(A, b)

line1 = w[2]*x**2 + w[1]*x + w[0]

plt.plot(x, f, 'r', label='real')
plt.plot(x, line1, 'g', label='approximation')
plt.legend(prop={'size':10})
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

# Приближаем третьей степенью

```python
X = np.array([1, 4, 10, 15])

A = np.c_[np.ones(4), X, X**2, X**3]
b = np.sin(X/5) * np.exp(X/10) + 5*np.exp(-X/2)

w = np.linalg.solve(A, b)

line1 = w[3]*x**3 + w[2]*x**2 + w[1]*x + w[0]

plt.plot(x, f, 'r', label='real')
plt.plot(x, line1, 'g', label='approximation')
plt.legend(prop={'size':10})
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
