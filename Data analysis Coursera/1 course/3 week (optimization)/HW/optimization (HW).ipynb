# Optimization homework (finding the minimum)

## â„–1 (smooth function)

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import optimize

# define our func to use it in our algorithms
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

# plotting the func
x = np.linspace(1, 30, 100)
func = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)

plt.plot(x, func)
plt.title("Function for optimization")
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# finding the minimum from different points
bfgs_min1 = optimize.minimize(f, [2], method='BFGS')
print(round(bfgs_min1.fun, 2))

bfgs_min2 = optimize.minimize(f, [30], method='BFGS')
print(bfgs_min2)

# writing the answer to a file
with open("answer_task1.txt", "w") as f:
    f.write(f"{round(bfgs_min1.fun, 2)} {round(bfgs_min2.fun, 2)}")
```

## â„–2 (genetic algorithm)

```python
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

bounds = [(0, 30)]
global_min = optimize.differential_evolution(f, bounds)
print(global_min)

# writing the answer to a file
with open("answer_task2.txt", "w") as f:
    f.write(str(round(float(global_min.x), 2)))   
    
```

## â„–3 (non-smooth function)

```python
# plotting the func
x = np.linspace(1, 30, 100)
h_func = (np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)).astype(int)

plt.plot(x, h_func)
plt.title("Non-smooth function for optimization")
plt.xlabel('x')
plt.ylabel('y')
plt.show()


# define the functions to use it in the algorithms
def f(x):
    return math.sin(x / 5) * math.exp(x / 10) + 5 * math.exp(-x / 2)

def h(x):
    return int(f(x))

# BFGS method
print("BFGS method")

bfgs_min = optimize.minimize(h, [30], method = 'BFGS')
print(bfgs_min)

# genetic method
print("\nDifferential evolution method")

bounds = [(0,30)]
gen_min = optimize.differential_evolution(h, bounds)
print(gen_min)

# writing the answer to a file
with open("answer_task3.txt", "w") as f:
    f.write(f"{bfgs_min.fun} {gen_min.fun}")
    
```

**Ð¡onclusion:** we made sure that the BFGS method can't find the min point of non-smooth function, because this method is gradient.
